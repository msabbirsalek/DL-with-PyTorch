```python
import csv
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams['figure.facecolor'] = '#ffffff'   # Background_color = white (fot matplotlib figures)
```

```python
def from_csv(filename):
    # initializing the titles and rows list
    fields = []
    rows = []
    
    # reading csv file
    with open(filename, 'r') as csvfile:
        # creating a csv reader object
        csvreader = csv.reader(csvfile)
      
        # extracting field names through first row
        fields = next(csvreader)
  
        # extracting each data row one by one
        for row in csvreader:
            rows.append(row)
            
    return rows
```

```python
model0_losses = from_csv('/HW_1_1_CIFAR10_model0_train_losses.csv')
model1_losses = from_csv('/HW_1_1_CIFAR10_model1_train_losses.csv')
model2_losses = from_csv('/HW_1_1_CIFAR10_model2_train_losses.csv')

model0_losses = [float(x[0]) for x in model0_losses]
model1_losses = [float(x[0]) for x in model1_losses]
model2_losses = [float(x[0]) for x in model2_losses]
```

```python
def plot_losses(model0_losses,model1_losses,model2_losses):
    plt.plot(model0_losses, label="Model_0")
    plt.plot(model1_losses, label="Model_1")
    plt.plot(model2_losses, label="Model_2")
    plt.xlabel('Epoch')
    plt.xticks([x for x in range(0,21,2)])
    plt.ylabel('Loss')
    plt.legend(loc="upper right")
    plt.title('Loss vs. No. of epochs')
    plt.show();
```

```python
plot_losses(model0_losses,model1_losses,model2_losses)
```

```python
model0_train_acc = from_csv('/HW_1_1_CIFAR10_model0_train_accuracies.csv')
model1_train_acc = from_csv('/HW_1_1_CIFAR10_model1_train_accuracies.csv')
model2_train_acc = from_csv('/HW_1_1_CIFAR10_model2_train_accuracies.csv')

model0_val_acc = from_csv('/HW_1_1_CIFAR10_model0_val_accuracies.csv')
model1_val_acc = from_csv('/HW_1_1_CIFAR10_model1_val_accuracies.csv')
model2_val_acc = from_csv('/HW_1_1_CIFAR10_model2_val_accuracies.csv')

model0_train_acc = [float(x[0]) for x in model0_train_acc]
model1_train_acc = [float(x[0]) for x in model1_train_acc]
model2_train_acc = [float(x[0]) for x in model2_train_acc]

model0_val_acc = [float(x[0]) for x in model0_val_acc]
model1_val_acc = [float(x[0]) for x in model1_val_acc]
model2_val_acc = [float(x[0]) for x in model2_val_acc]
```

```python
def plot_accuracies(model0_train_acc, model1_train_acc, model2_train_acc,
                    model0_val_acc, model1_val_acc, model2_val_acc):
    plt.plot(model0_train_acc, '-x', label="Model_0_train")
    plt.plot(model0_val_acc, '-x', label="Model_0_test")
    plt.plot(model1_train_acc, '-x', label="Model_1_train")
    plt.plot(model1_val_acc, '-x', label="Model_1_test")
    plt.plot(model2_train_acc, '-x', label="Model_2_train")
    plt.plot(model2_val_acc, '-x', label="Model_2_test")
    plt.xlabel('Epoch')
    plt.xticks([x for x in range(0,21,2)])
    plt.ylabel('Accuracy')
    plt.legend(loc="lower right")
    plt.title('Accuracy vs. No. of epochs')
    plt.show();
```

```python
plot_accuracies(model0_train_acc, model1_train_acc, model2_train_acc,
                    model0_val_acc, model1_val_acc, model2_val_acc)
```
